## Handling projections

Let's project this dataframe to match the CRS of the itslive dataset

```python
#project rgi data to match itslive
se_asia_prj = se_asia.to_crs('EPSG:32645') #we know the epsg from looking at the 'spatial epsg' attr of the mapping var of the dc object
se_asia_prj.head(3)
```

Give each glacier (row) a unique integer key that is related to that glacier's RGIId. We will use this later

```python
se_asia_prj['RGI_int'] = se_asia_prj['RGIId'].str.slice(9,).replace('.','_')
se_asia_prj['RGI_int'] = se_asia_prj.RGI_int.apply(lambda x: int('15' + x))
se_asia_prj.RGI_int
```

Subset only glaciers larger than 5km2:

```python
se_asia_prj = se_asia_prj.loc[se_asia_prj['Area'] > 5.]
se_asia_prj.head()
```

Next, want to subset the RGI dataset by the spatial extent of the ITS_LIVE data.
First, get the bbox of the ITS_LIVE data as a vector

```python
dc_bbox = itslivetools.get_bbox_single(dc)
```

Project it to local UTM to match the RGI geodataframe and extract the coordinate values from the geometry column

```python
dc_bbox_prj = dc_bbox.to_crs('EPSG:32645')
```

Subset RGI dataset:

```python
rgi_sub = gpd.sjoin(se_asia_prj, dc_bbox_prj, how='inner')
# need to set the type as string here bc for some reason its object intead of str
rgi_sub["RGIId"] = rgi_sub.RGIId.astype("string")
rgi_sub.head()
```

```python
#now, get velocity statistics of each 'region' (integer) using the mask
grouped_ID = out_grid_se_asia.drop('spatial_ref').groupby(out_grid_se_asia.RGI_int)
grouped_ID
```

```python
grid_mean_sp = grouped_ID.mean().rename({'speed': 'speed_mean'})
grid_median_sp = grouped_ID.median().rename({'speed': 'speed_median'})
grid_min_sp = grouped_ID.min().rename({'speed': 'speed_min'})
grid_max_sp = grouped_ID.max().rename({'speed': 'speed_max'})
grid_std_sp = grouped_ID.max().rename({'speed': 'speed_std'})
```

```python
zonal_stats = xr.merge([grid_mean_sp, grid_median_sp, grid_min_sp, grid_max_sp, grid_std_sp]).to_dataframe()
zonal_stats = zonal_stats.reset_index()
zonal_stats
```

```python
#now, trying to merge zonal stats df back with original glacier df on integer_ID col
se_asia_glacier_data = se_asia_prj.loc[se_asia_prj['Area'] > 5.].merge(zonal_stats, on='RGI_int')
```
